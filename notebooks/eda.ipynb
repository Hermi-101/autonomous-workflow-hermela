```python
query = """
SELECT 
    c.id_employee, 
    c.timestamp, 
    e.app_name, 
    e.window_title, 
    e.event, 
    e.url, 
    c.image_path
FROM captures c
JOIN events e ON c.event_id = e.id
ORDER BY c.id_employee, c.timestamp;
"""
df = pd.read_sql(query, conn)


print(df["id_employee"].unique())
```

```python
query = """
SELECT 
    c.id_employee, 
    c.timestamp, 
    e.app_name, 
    e.window_title, 
    e.event, 
    e.url, 
    c.image_path
FROM captures c
JOIN events e ON c.event_id = e.id
ORDER BY c.id_employee, c.timestamp;
"""

df = pd.read_sql(query, conn)

df["timestamp"] = pd.to_datetime(df["timestamp"], format='ISO8601')

stats = df.groupby("id_employee")["timestamp"].agg(['min', 'max'])
print(stats)
```
